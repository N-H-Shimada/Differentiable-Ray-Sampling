```python
path = './save/all-instance-car_30000itr/'
```

```python
# Lossのplot
def Plot_loss() :
    m = 500
    
    loss_train_list = []
    with open(path+ 'loss_train.txt') as f:
        data_temp = f.read()
        data_temp = data_temp.split('\n')
    length = int(len(data_temp) / m)
    for i in range(length):
        temp_sum = 0.0
        for j in range(m):
            index = i*m + j
            temp_sum += float(data_temp[index])
        temp_sum = temp_sum / m
        loss_train_list.append(temp_sum)
        
    loss_valid_list = []
    with open(path+ 'loss_valid.txt') as f:
        data_temp = f.read()
        data_temp = data_temp.split('\n')
    length = int(len(data_temp) / m)
    for i in range(length):
        temp_sum = 0.0
        for j in range(m):
            index = i*m + j
            temp_sum += float(data_temp[index])
        temp_sum = temp_sum / m
        loss_valid_list.append(temp_sum)   
    
    train_time = (np.arange(length) + 1) * m
    
    plt.figure(figsize=(6, 4), dpi=120)
    #plt.xlim(1,length)
    plt.xlim(-1000,30000)
    plt.ylim(0.05,0.12)
    plt.plot(train_time, loss_train_list, '-o', label='Train')
    plt.plot(train_time, loss_valid_list, '--', label='Valid')
    #plt.savefig('./figure/Loss' + '.png')
    plt.legend()
    plt.show()
    
Plot_loss()
```

```python
def delete():
    plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
    plt.tick_params(bottom=False, left=False, right=False, top=False)

def Func(name, mode_index, list_index, num_sample): 
    
    # load DRS
    drs = DRS(w)
    chainer.serializers.load_npz(path + name, drs)
    
    # get path
    mode = ['Train', 'Valid']
    mode = mode[mode_index]
    name_list = [Get_name_list(mode)[list_index]]
    print(name_list)
    
    for i in range(5):
        # data load
        A, B, C, D = Data_loader(name_list, 1, 224*224*(i+1), h, w, h_image, w_image, -1, mode)
        
        # Forward
        output = drs(A, B, C, num_sample) # [bs, 224*224]

        # Plot (RGB, Ground truth Mask, Output Mask)
        D_show =  D[0].reshape(224,224)
        output_show = output[0].data.reshape(224,224)
        diff = np.ones((224, 224, 3))  # B = 1.0
        diff[:,:,0] = D_show           # R = 0.0 (if masked in Ground Truth)
        diff[:,:,1] = output_show      # G = 0.0 (if masked in Pred)
        
#         if (i==5):
#             print('----------------------- test -------------------------')
        fig = plt.figure(figsize=(10, 10))
        plt.subplot(1, 4, 1)
        delete()
        plt.imshow(C[0].transpose(1,2,0))
        plt.subplot(1, 4, 2)
        delete()
        plt.imshow(D_show, cmap='gray', vmin=0.0, vmax=1.0)
        plt.subplot(1, 4, 3)
        delete()
        plt.imshow(output_show, cmap='gray', vmin=0.0, vmax=1.0)
        plt.subplot(1, 4, 4)
        delete()
        plt.imshow(diff, cmap='gray', vmin=0.0, vmax=1.0)
        plt.show()
```
